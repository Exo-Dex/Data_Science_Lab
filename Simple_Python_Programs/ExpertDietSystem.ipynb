## Nutritional Guidelines and Daily Calorie Intake Calculation

To establish a knowledge base of nutritional guidelines, we will define the following:

### 1. Daily Calorie Intake Calculation Methods

We will use common formulas to estimate daily calorie needs based on individual factors. These formulas typically involve calculating the Basal Metabolic Rate (BMR) and then applying an activity factor.

*   **Basal Metabolic Rate (BMR) Formulas:**
    *   **Mifflin-St Jeor Equation (Recommended for most adults):**
        *   For men: `BMR = (10 * weight in kg) + (6.25 * height in cm) - (5 * age in years) + 5`
        *   For women: `BMR = (10 * weight in kg) + (6.25 * height in cm) - (5 * age in years) - 161`

    *   **Harris-Benedict Equation (Older, but still used):**
        *   For men: `BMR = 66.5 + (13.75 * weight in kg) + (5.003 * height in cm) - (6.755 * age in years)`
        *   For women: `BMR = 655.1 + (9.563 * weight in kg) + (1.850 * height in cm) - (4.676 * age in years)`

*   **Activity Factor (to calculate Total Daily Energy Expenditure - TDEE):**
    *   Sedentary (little or no exercise): `TDEE = BMR * 1.2`
    *   Lightly active (light exercise/sports 1-3 days/week): `TDEE = BMR * 1.375`
    *   Moderately active (moderate exercise/sports 3-5 days/week): `TDEE = BMR * 1.55`
    *   Very active (hard exercise/sports 6-7 days a week): `TDEE = BMR * 1.725`
    *   Extra active (very hard exercise/physical job): `TDEE = BMR * 1.9`

### 2. Healthy Macronutrient Ratios

Recommended macronutrient distribution ranges (AMDRs) are typically expressed as percentages of total daily calorie intake.

*   **Carbohydrates:** 45-65% of total daily calories
*   **Fats:** 20-35% of total daily calories (with saturated fats <10%)
*   **Proteins:** 10-35% of total daily calories

These guidelines will be used to inform the expert system's recommendations.

## Collect User Dietary Preferences

### Subtask:
Develop a mechanism to collect essential user information such as age, gender, activity level, dietary restrictions (e.g., vegetarian, vegan, allergies), and specific health goals (e.g., weight loss, muscle gain).


**Reasoning**:
The subtask is to develop a mechanism to collect essential user information. This code block will prompt the user for their age, weight, height, gender, activity level, dietary restrictions, and health goal, storing all inputs in a dictionary as instructed.



```python
user_info = {}

# 1. Prompt the user for age, weight, and height
user_info['age'] = int(input("Enter your age (in years): "))
user_info['weight'] = float(input("Enter your weight (in kg): "))
user_info['height'] = float(input("Enter your height (in cm): "))

# 2. Ask the user to select their gender
while True:
    gender = input("Enter your gender ('male' or 'female'): ").lower()
    if gender in ['male', 'female']:
        user_info['gender'] = gender
        break
    else:
        print("Invalid input. Please enter 'male' or 'female'.")

# 3. Ask the user to select their activity level
activity_levels = ['sedentary', 'lightly active', 'moderately active', 'very active', 'extra active']
while True:
    activity = input(f"Select your activity level from {activity_levels}: ").lower()
    if activity in activity_levels:
        user_info['activity_level'] = activity
        break
    else:
        print(f"Invalid input. Please choose from {activity_levels}.")

# 4. Prompt the user for dietary restrictions or allergies
dietary_restrictions_input = input("Enter any dietary restrictions or allergies (e.g., 'vegetarian', 'vegan', 'gluten-free', 'none'). Separate multiple entries with commas: ")
user_info['dietary_restrictions'] = [item.strip().lower() for item in dietary_restrictions_input.split(',') if item.strip() != 'none']

# 5. Ask the user to specify their health goal
health_goals = ['weight loss', 'maintain weight', 'muscle gain']
while True:
    goal = input(f"Specify your health goal from {health_goals}: ").lower()
    if goal in health_goals:
        user_info['health_goal'] = goal
        break
    else:
        print(f"Invalid input. Please choose from {health_goals}.")

print("User information collected successfully:")
print(user_info)
```

## Implement Decision-Making Logic

### Subtask:
Based on the defined nutritional guidelines and collected user preferences, create a set of rules or a decision tree to recommend appropriate food items and meal components for different meals (breakfast, lunch, dinner, snacks).


**Reasoning**:
The first few instructions involve defining functions to calculate BMR, TDEE, adjusted TDEE based on health goals, and macronutrient targets. These are foundational calculations for the diet plan generation, so I will implement them in the first code block.



```python
def calculate_bmr(user_info):
    weight = user_info['weight']
    height = user_info['height']
    age = user_info['age']
    gender = user_info['gender']

    if gender == 'male':
        bmr = (10 * weight) + (6.25 * height) - (5 * age) + 5
    else: # female
        bmr = (10 * weight) + (6.25 * height) - (5 * age) - 161
    return bmr

def calculate_tdee(bmr, activity_level):
    activity_factors = {
        'sedentary': 1.2,
        'lightly active': 1.375,
        'moderately active': 1.55,
        'very active': 1.725,
        'extra active': 1.9
    }
    return bmr * activity_factors.get(activity_level, 1.2)

def adjust_tdee_for_goal(tdee, health_goal):
    if health_goal == 'weight loss':
        return tdee - 500  # Calorie deficit
    elif health_goal == 'muscle gain':
        return tdee + 300  # Calorie surplus
    else: # maintain weight
        return tdee

def calculate_macronutrient_targets(daily_calories):
    # Macronutrient ratios from guidelines
    carb_min_percent, carb_max_percent = 0.45, 0.65
    fat_min_percent, fat_max_percent = 0.20, 0.35
    protein_min_percent, protein_max_percent = 0.10, 0.35

    # Calories per gram
    calories_per_gram_protein = 4
    calories_per_gram_carbs = 4
    calories_per_gram_fat = 9

    # Calculate target ranges for each macronutrient in grams
    macro_targets = {
        'Carbohydrates': {
            'min_grams': (daily_calories * carb_min_percent) / calories_per_gram_carbs,
            'max_grams': (daily_calories * carb_max_percent) / calories_per_gram_carbs
        },
        'Fats': {
            'min_grams': (daily_calories * fat_min_percent) / calories_per_gram_fat,
            'max_grams': (daily_calories * fat_max_percent) / calories_per_gram_fat
        },
        'Proteins': {
            'min_grams': (daily_calories * protein_min_percent) / calories_per_gram_protein,
            'max_grams': (daily_calories * protein_max_percent) / calories_per_gram_protein
        }
    }
    return macro_targets

# Example usage with user_info (assuming user_info is already defined)
bmr = calculate_bmr(user_info)
tdee = calculate_tdee(bmr, user_info['activity_level'])
adjusted_daily_calories = adjust_tdee_for_goal(tdee, user_info['health_goal'])
macronutrient_targets = calculate_macronutrient_targets(adjusted_daily_calories)

print("Calculated BMR:", round(bmr, 2), "calories")
print("Calculated TDEE:", round(tdee, 2), "calories")
print("Adjusted Daily Calorie Target:", round(adjusted_daily_calories, 2), "calories")
print("Macronutrient Targets (grams):")
for macro, values in macronutrient_targets.items():
    print(f"  {macro}: {round(values['min_grams'], 2)} - {round(values['max_grams'], 2)} g")

print("Nutritional targets calculated.")
```

**Reasoning**:
I need to implement the next instruction, which is to create a function to filter the food database based on user's dietary restrictions. This function will iterate through the food database and remove items that do not comply with the specified restrictions.



```python
def filter_food_database(food_db, user_restrictions):
    filtered_db = []
    for food_item in food_db:
        is_compliant = True
        item_category = food_item['category'].lower()
        item_name = food_item['item'].lower()

        for restriction in user_restrictions:
            if restriction == 'vegetarian':
                # Vegetarian excludes meat (Proteins from animals like chicken, salmon)
                # Assume Dairy, Eggs, Plant-based proteins are okay.
                if item_category == 'proteins' and not (item_name.find('egg') != -1 or item_name.find('tofu') != -1 or item_name.find('lentil') != -1 or item_name.find('bean') != -1 or item_name.find('yogurt') != -1):
                    if 'chicken' in item_name or 'salmon' in item_name or 'beef' in item_name or 'pork' in item_name:
                        is_compliant = False
                        break
            elif restriction == 'vegan':
                # Vegan excludes all animal products: meat, dairy, eggs, honey
                if item_category in ['proteins', 'dairy']:
                    if 'chicken' in item_name or 'salmon' in item_name or 'beef' in item_name or 'pork' in item_name or 'yogurt' in item_name or 'egg' in item_name:
                        is_compliant = False
                        break
                # Also check for non-categorized items that might be animal products
                if 'honey' in item_name: # Example for other animal products
                    is_compliant = False
                    break
            elif restriction == 'gluten-free':
                # Gluten-free excludes wheat, barley, rye
                if 'wheat' in item_name or 'barley' in item_name or 'rye' in item_name or 'bread' in item_name or 'pasta' in item_name or 'oats' in item_name and 'gluten-free' not in item_name: # Simple check, can be expanded
                    is_compliant = False
                    break
            # Add more restrictions as needed

        if is_compliant:
            filtered_db.append(food_item)

    return filtered_db

# Filter the food database based on user info
filtered_food_database = filter_food_database(food_database, user_info['dietary_restrictions'])

print(f"Original food database size: {len(food_database)}")
print(f"Filtered food database size: {len(filtered_food_database)}")
print("First 5 items in filtered food database:")
for i in range(min(5, len(filtered_food_database))):
    print(filtered_food_database[i]['item'])
```

**Reasoning**:
The next instruction is to outline a basic structure for meal components for different meals. I will create a dictionary to define these structures.



```python
meal_structures = {
    'breakfast': ['Proteins', 'Carbohydrates', 'Fruits', 'Dairy'],
    'lunch': ['Proteins', 'Carbohydrates', 'Vegetables', 'Fats'],
    'dinner': ['Proteins', 'Carbohydrates', 'Vegetables', 'Fats'],
    'snack': ['Proteins', 'Carbohydrates', 'Fruits', 'Fats']
}

print("Meal structures defined successfully:")
for meal, components in meal_structures.items():
    print(f"  {meal.capitalize()}: {', '.join(components)}")
```

**Reasoning**:
Now that the nutritional targets, meal structures, and filtered food database are available, the next step is to create a function that intelligently recommends food items for each meal based on these parameters. This will involve iterating through meal structures and selecting appropriate foods from the filtered database, while considering the macronutrient targets to generate a balanced meal plan.



```python
import random

def generate_meal_plan(adjusted_daily_calories, macronutrient_targets, filtered_food_database, meal_structures):
    meal_plan = {}
    remaining_calories = adjusted_daily_calories

    # Helper function to get food items by category
    def get_foods_by_category(category):
        return [item for item in filtered_food_database if item['category'] == category]

    for meal_name, components in meal_structures.items():
        meal_plan[meal_name] = []
        meal_calories = 0 # Track calories for the current meal
        meal_macros = {'protein': 0, 'carbohydrates': 0, 'fat': 0}

        for component_category in components:
            available_foods = get_foods_by_category(component_category)
            if not available_foods:
                continue

            # Randomly select a food item from the category
            chosen_food = random.choice(available_foods)

            # For simplicity, we'll add one serving of the chosen food
            # In a more advanced system, serving sizes would be adjusted to meet macro targets more precisely
            meal_plan[meal_name].append({
                'item': chosen_food['item'],
                'category': chosen_food['category'],
                'serving_size_g': chosen_food['serving_size_g'],
                'calories': chosen_food['calories_per_serving'],
                'protein_g': chosen_food['protein_g_per_serving'],
                'carbohydrates_g': chosen_food['carbohydrates_g_per_serving'],
                'fat_g': chosen_food['fat_g_per_serving']
            })
            meal_calories += chosen_food['calories_per_serving']
            meal_macros['protein'] += chosen_food['protein_g_per_serving']
            meal_macros['carbohydrates'] += chosen_food['carbohydrates_g_per_serving']
            meal_macros['fat'] += chosen_food['fat_g_per_serving']

        meal_plan[meal_name].insert(0, {'meal_total_calories': meal_calories, 'meal_total_protein_g': meal_macros['protein'], 'meal_total_carbohydrates_g': meal_macros['carbohydrates'], 'meal_total_fat_g': meal_macros['fat']})

    return meal_plan

# Generate the meal plan
generated_plan = generate_meal_plan(adjusted_daily_calories, macronutrient_targets, filtered_food_database, meal_structures)

print("Generated Meal Plan:")
total_plan_calories = 0
total_plan_protein = 0
total_plan_carbs = 0
total_plan_fat = 0

for meal, items in generated_plan.items():
    print(f"\n--- {meal.upper()} ---")
    for i, item_data in enumerate(items):
        if i == 0:
            print(f"  Total Meal Calories: {round(item_data['meal_total_calories'], 2)}")
            print(f"  Total Meal Protein: {round(item_data['meal_total_protein_g'], 2)}g")
            print(f"  Total Meal Carbohydrates: {round(item_data['meal_total_carbohydrates_g'], 2)}g")
            print(f"  Total Meal Fat: {round(item_data['meal_total_fat_g'], 2)}g")
            total_plan_calories += item_data['meal_total_calories']
            total_plan_protein += item_data['meal_total_protein_g']
            total_plan_carbs += item_data['meal_total_carbohydrates_g']
            total_plan_fat += item_data['meal_total_fat_g']
        else:
            print(f"  - {item_data['item']} ({item_data['serving_size_g']}g): {item_data['calories']} calories, {item_data['protein_g']}g P, {item_data['carbohydrates_g']}g C, {item_data['fat_g']}g F")

print("\n--- DAILY TOTALS ---")
print(f"Target Daily Calories: {round(adjusted_daily_calories, 2)}")
print(f"Actual Daily Calories: {round(total_plan_calories, 2)}")
print(f"Target Protein: {round(macronutrient_targets['Proteins']['min_grams'], 2)} - {round(macronutrient_targets['Proteins']['max_grams'], 2)}g")
print(f"Actual Protein: {round(total_plan_protein, 2)}g")
print(f"Target Carbohydrates: {round(macronutrient_targets['Carbohydrates']['min_grams'], 2)} - {round(macronutrient_targets['Carbohydrates']['max_grams'], 2)}g")
print(f"Actual Carbohydrates: {round(total_plan_carbs, 2)}g")
print(f"Target Fats: {round(macronutrient_targets['Fats']['min_grams'], 2)} - {round(macronutrient_targets['Fats']['max_grams'], 2)}g")
print(f"Actual Fats: {round(total_plan_fat, 2)}g")

print("Meal plan generated and printed.")
```
