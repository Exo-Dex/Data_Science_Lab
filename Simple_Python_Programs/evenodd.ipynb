```python
num_str = input("Enter an integer: ")

try:
    num = int(num_str)
    if num % 2 == 0:
        print(f"The number {num} is even.")
    else:
        print(f"The number {num} is odd.")
except ValueError:
    print(f"Invalid input: '{num_str}' is not a valid integer.")
```
