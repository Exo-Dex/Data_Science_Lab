```python
titanic_data = pd.read_csv("/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv")
titanic_data.head()
```

```python
sns.swarmplot(x=titanic_data['Sex'],
              y=titanic_data['Age'],
             hue=titanic_data['Survived'])
plt.title("SwarmPlot Survival Distribution: Age vs Sex")
```

```python
sns.countplot(x='Embarked', hue='Survived', data=titanic_data)
plt.title('Survival Counts by Embarked')
```

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Survival Counts by Passanger Class')
```

```python
sns.swarmplot(data=titanic_data, y='Fare', x='Pclass', hue='Survived')
plt.title("Survival Distribution With relation to 'Pclass vs Fare'")
```
