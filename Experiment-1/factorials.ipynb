This program will calculate the factorial of a non-negative integer entered by the user. The factorial of a non-negative integer `n` is the product of all positive integers less than or equal to `n`.

```python
# Function to calculate factorial
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact = fact * i
        return fact

# Get input from the user
try:
    num = int(input("Enter a non-negative integer: "))
    result = factorial(num)
    print(f"The factorial of {num} is {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```
